#### **2. Écrire une fonction prenant comme argument un entier n et simulant n pas de marche aléatoire puis affichant la position finale. Modifier le programme pour qu’il affiche la liste de toutes les positions intermédiaires et les représente graphiquement. Vous afficherez en abscisse le numéro de l’étape et en ordonnée la position de l’objet à cette étape. Tester votre programme avec différentes valeurs de n**

```python
import matplotlib.pyplot as plt
import numpy as np

def random_etape():
    return np.random.choice([-1, 1])

def marche_aleatoire(n):
    positions = [0]
    for _ in range(n):
        etape = random_etape()
        positions.append(positions[-1] + etape)

    print(f"Position finale après {n} étapes: {positions[-1]}")
    print(f"Positions intermédiaires: {positions}")

    # Plot the positions
    plt.plot(positions)
    plt.xlabel("Numéro de l'etape")
    plt.ylabel('Position')
    plt.title(f'Marche aléatoire de {n} étapes')
    plt.show()

values = [10**2, 10**4]
for n in values:
    print(f"Simulation pour n = {n}")
    marche_aleatoire(n)
```

#### **3. Modifier votre programme pour qu’il calcule le temps moyen de passage entre deux visites de l’état initial. Par exemple, si la séquence des états est la suivante : [0, −1, 0, 1, 2, 3, 2, 1, 0], les temps de passage entre 2 visites de l’état 0 sont 2 et 6, ce qui donne une moyenne de 4. Tester pour différentes valeurs de n. Commenter. Petit challenge : réaliser cette simulation avec le plus grand n possible en moins de 1 seconde.**


```python
import time

def random_etape():
    return np.random.choice([-1, 1])

def marche_aleatoire(n):
    positions = [0]
    passages = 0
    start_time = time.time()  # Registra o tempo de início

    for _ in range(n):
        etape = random_etape()
        positions.append(positions[-1] + etape)

        if (positions[-2], positions[-1]) in [(0, 1), (1, 0)]:
            passages += 1

    end_time = time.time()
    duration = end_time - start_time

    # Exibe o tempo de execução
    print(f"Duration pour {n} etapes: {duration}")

    return passages, duration

for exponent in range(3, 7):
  n = 10 ** exponent
  print(f"Testing with n = {n}")
  _, duration = marche_aleatoire(n)
```

#### **4. Ecrire une fonction qui estime la probabilité que l’objet soit en position k après n étapes, c’est-à-dire P(Xn = k), en effectuant m simulations ***
Vérifier que vos résultats empiriques sont cohérents avec cette formule
pour plusieurs valeurs de n et k, en augmentant progressivement la
valeur de m

```python
import numpy as np
import math

def estimer_probabilite(n, k, m):
  # Vérifier si k et n ont la même parité
  if (n + k) % 2 != 0:
    return 0

  etapes = np.random.choice([-1, 1], size=(m, n))
  final_positions = etapes.sum(axis=1)

  freq_k = np.count_nonzero(final_positions == k)
  prob = freq_k / m

  return prob



probabilite_estimee1 = estimer_probabilite(30, 10, 100)
probabilite_estimee2 = estimer_probabilite(30, 10, 10000)

probabilite_theorique = (1 / (2 ** 30)) * math.comb(30, (30 + 10) // 2)

print(f"Probabilité estimée 1 : {probabilite_estimee1:.2f} | Probabilité estimée 2 : {probabilite_estimee2:.2f}")
print(f"Probabilité théorique : {probabilite_theorique:.2f}")
```

#### **5. Une autre méthode, pour calculer les probabilités en régime transitoire de manière exacte, est d’utiliser la relation de récurrence suivante, vue en cours.** Utiliser cette relation, dans le cadre d’une marche aléatoire 1D, pour écrire une fonction prenant en entrée k et n et retournant P(Xn = k). Vérifier que vos résultats sont cohérents avec la formule (1).

```python
import numpy as np
import math

def probabilite_marche_aleatoire(n, k):
  if n == 0:
    return 1 if k == 0 else 0

  if k < -n or k > n or (n + k) % 2 != 0:
    return 0

  return 0.5 * (probabilite_marche_aleatoire(n-1, k-1) +
                  probabilite_marche_aleatoire(n-1, k+1))

k = 2
n = 10

probabilite_theorique = (1 / (2 ** n)) * math.comb(n, (n + k) // 2)



print(f"P(X_{n} = {k}) = {probabilite_marche_aleatoire(n, k):.4f}")
print(f"Theorique P(X_{n} = {k}) = {probabilite_theorique:.4f}")
```
